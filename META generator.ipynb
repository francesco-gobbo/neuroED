```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from pandas.api.types import CategoricalDtype

BEH_directory=r'C:\Users\fgobbo\OneDrive - University of Edinburgh\arena room files\ANALYSIS 2.0\Data\H2234\Allo\DLC'

#Read directory containing all beh vids and read in Meta.csv file loc
#BEH_directory = folder_name # What is the address of the directory containing videos 
#Meta_csv = pd.read_csv(os.path.join(BEH_directory,'h2230 allo ego LED.csv'))
Meta = pd.DataFrame({'Animal':'',
                        'Allo/Ego':'',
                        'Session':'',
                        'Stage':'',
                        'Trial':'',
                        'Combo':'', 
                        'Sandwell':'' 
                    }, index=[0])


for file in os.listdir(BEH_directory):
    if file[-4:]=='.csv':
        animal, session, combo_sandwell, stage, dlc = (file[:-4].split("_"))
        allo_ego = session[0]; session = session[1:]
        combo=combo_sandwell[:2]; sandwell=combo_sandwell[-1:]
        stage_part = stage[:-1]; trial = stage[-1:]
        print(animal, allo_ego, session, combo, sandwell,stage_part, trial)

        # check to see if trial data is already in meta


        add_meta = pd.DataFrame({'Animal':animal,
                            'Allo/Ego':allo_ego,
                            'Session':session,
                            'Stage':stage_part,
                            'Trial':trial,
                            'Combo':combo, 
                            'Sandwell':sandwell 
                        }, index=[0])
        Meta = Meta.append(add_meta)

order = CategoricalDtype( ['PRE', 'SAM', 'CHO', 'CTR', 'CTC', 'POS'], ordered=True)
        
Meta_csv= Meta.reset_index(drop=True)
Meta_csv=Meta_csv.drop([0])
Meta_csv['Stage'] = Meta_csv['Stage'].astype(order)
Meta_csv=Meta_csv.sort_values(['Session','Stage', 'Trial'])

Meta_csv.to_csv(os.path.join(BEH_directory, 'H2230_META.csv'), mode='w', header=True, index=False)
    
        #Isolate meta info

Meta_csv
    
#Save file                         
#Meta_csv.to_csv(os.path.join(folder_name,'meta', 'Meta.csv'), mode='w', header=True, index=False)
```
